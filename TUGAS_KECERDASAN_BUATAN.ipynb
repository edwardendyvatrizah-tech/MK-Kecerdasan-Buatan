```python
import numpy as np
import matplotlib.pyplot as plt

# Fungsi parameter
def triangle(x, a, b, c, d, e):
    """
    Calculates the membership value for a triangular fuzzy function using if/elif/else.
    Parameters:
    x (float or np.array): The input value(s).
    a (float): The left-most X point, paired with y value d.
    b (float): The middle X point, paired with Y value of 1.
    c (float): The right-most X point, paired with Y value e.
    d (float): The y-value at point a.
    e (float): The y-value at point c.
    Returns:
    float or np.array: The membership value(s).
    """
    if x <= a:
        return 1
    if x <= b:
        return 1
    elif b < x <= c:
        return (c - x) / (c - b)
    else:
        return 0

x = np.linspace(0, 40, 100)
y = np.zeros_like(x)

for i, x_val in enumerate(x):
       y[i] = triangle(x_val, 0, 15, 25, 1, 0)

# 2. Plotting
plt.figure(figsize=(8, 4))
plt.plot(x, y, color='red', linewidth=2)

# Menambahkan Label "Dingin"
plt.text(14, 1.05, 'Dingin', color='red', fontsize=12, fontweight='bold', ha='center')

# Pengaturan Axis (Sumbu)
plt.xlabel('Temperatur $t$', fontsize=12)
plt.ylabel('$\\mu_{tD}(t)$', rotation=0, labelpad=20, fontsize=12)
plt.ylim(-0.1, 1.2)
plt.xlim(10, 38)

# Mengatur angka yang muncul disumbu
plt.xticks([20, 30])
plt.yticks([0, 1])

plt.grid(False)
plt.show()
```
