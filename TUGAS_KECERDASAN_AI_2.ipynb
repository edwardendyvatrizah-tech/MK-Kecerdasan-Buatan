```python
import numpy as np
import matplotlib.pyplot as plt

def triangle (x,a,b,c,d,e):
    """
    Calculates the membership value for a triangular fuzzy function using if/elif/else.

    Parameters:
    x (float or np.array): The input value(s).
    a (float): The left-most X point, paired with y value d.
    b (float): The middle X point, paired with Y value of 1.
    c (float): The right-most X point, paired with Y value e.
    d (float): The y-value at point a.
    e (float): The y-value at point c.

    Returns:
    float or np.array: The membership value(s).
    """

    if x < a:
        return d
    elif a <= x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return (c - x) / (c - b)
    elif x > c:
        return e
    else:
        return 0

def fuzifikasi_area(area):
    Sedikit = triangle(area, 20, 20, 50, 1, 0)
    Sedang = triangle(area, 25, 50, 75, 0 ,0)
    Banyak = triangle(area, 50, 80, 80, 0, 1)
    return np.array([Sedikit, Sedang, Banyak])

semua_area = np.linspace(0, 100, 500)
semua_FD_fuzzy_area = np.zeros((len(semua_area), 3))

for i, area in enumerate(semua_area):
    semua_FD_fuzzy_area[i] = fuzifikasi_area(area)

plt.figure(figsize=(10, 6))
plt.plot(semua_area, semua_FD_fuzzy_area, label =['Sedikit', 'Sedang', 'Banyak'])
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def triangle(x, a, b, c, d, e):
    """
    Calculates the membership value for a triangular fuzzy function using if/elif/else.

    Parameters:
    x (float or np.array): The input value(s).
    a (float): The left-most X point, paired with y value d.
    b (float): The middle X point, paired with Y value of 1.
    c (float): The right-most X point, paired with Y value e.
    d (float): The y-value at point a.
    e (float): The y-value at point c.

    Returns:
    float or np.array: The membership value(s).
    """

    if x < a:
        return d
    elif a <= x <= b:
        return (x - a) / (b - a)
    elif b < x <= c:
        return (c - x) / (c - b)
    elif x > c:
        return e
    else:
        return 0

def fuzifikasi_area(area):
  Sedikit = triangle(area, 20, 20, 50, 1, 0)
  Sedang = triangle(area, 25, 50, 75, 0, 0)
  Banyak = triangle(area,50 , 80, 80, 0, 1)
  return np.array([Sedikit, Sedang, Banyak])

def fuzifikasi_daya(daya):
  kecil = triangle(daya, 0, 450, 900, 1, 0)
  sedang = triangle(daya, 450, 1500, 2500, 0, 0)
  besar = triangle(daya, 2000 , 3000, 3000, 0, 1)
  return np.array([kecil, sedang, besar])


input_area = 50
input_daya = 1500

fuzzy_area = fuzifikasi_area(input_area)
fuzzy_daya = fuzifikasi_daya(input_daya)
# 200 = Basic
# 500 = Premium
# 1000 = Platinum
rule_FD = np.array([[200,200,500],
                    [200, 500, 1000],
                    [500, 500,1000]])

num = 0
denum = 0
for i, val_area in enumerate(fuzzy_area):
  for j, val_daya in enumerate(fuzzy_daya):
    # Indent these lines correctly and use rule_FD for output values
    num = num + np.minimum(val_area, val_daya) * rule_FD[i][j]
    denum = denum + np.minimum(val_area, val_daya)
crisp_output = num/denum
print("Harga Paket")
print(crisp_output)
```
